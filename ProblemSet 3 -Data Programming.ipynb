```python
# Step 1:  Import the necessary libraries

import pandas as pd
import numpy as np
```

```python
# Step 2. Import the dataset from https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user.

dataset = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
```

```python
# Step 3. Assign it to a variable called users

users = pd.read_csv(dataset, delimiter='|')
```

```python
# Step 4. Discover what is the mean age per occupation

mean_age_per_occupation = users.groupby('occupation')['age'].mean()

print("Mean age per occupation:")
print(mean_age_per_occupation)
print("\n")
```

```python
# Step 5: Discover the Male ratio per occupation and sort it from most to least

male_ratio_per_occupation = users[users['gender'] == 'M'].groupby('occupation').size() / users.groupby('occupation').size()
male_ratio_per_occupation = male_ratio_per_occupation.sort_values(ascending=False)

print("Male ratio per occupation (sorted):")
print(male_ratio_per_occupation)
print("\n")
```

```python
# Step 6: For each occupation, calculate the minimum and maximum ages

min_max_age_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])

print("Minimum and maximum ages per occupation:")
print(min_max_age_per_occupation)
print("\n")
```

```python
# Step 7: For each combination of occupation and sex, calculate the mean age

mean_age_per_occupation_sex = users.groupby(['occupation', 'gender'])['age'].mean()
print("Mean age for each combination of occupation and sex:")
print(mean_age_per_occupation_sex)
print("\n")
```

```python
# Step 8: For each occupation, present the percentage of women and men

gender_counts_per_occupation = users.groupby(['occupation', 'gender']).size()
total_counts_per_occupation = users.groupby('occupation').size()
gender_percentage_per_occupation = (gender_counts_per_occupation / total_counts_per_occupation).unstack() * 100
gender_percentage_per_occupation.rename(columns={'F': 'Female', 'M': 'Male'}, inplace=True)

print("Percentage of male and female per occupation:")
print(gender_percentage_per_occupation)
```

```python
# Step 1. Import the necessary libraries
import pandas as pd
import numpy as np
```

```python
# Step 2: Import the dataset from the provided file
data = pd.read_csv('wind.txt', delim_whitespace=True, na_values='NaN')
```

```python
# Step 3: Replace the first 3 columns by a proper datetime index
data['Date'] = pd.to_datetime(data[['Yr', 'Mo', 'Dy']].astype(str).agg('-'.join, axis=1))
data = data.set_index('Date')
data = data.drop(['Yr', 'Mo', 'Dy'], axis=1)
data.head()
```

```python
# Step 4: Function to fix the year and apply it
def fix_year(x):
    if x.year > 2000:
        return x - pd.DateOffset(years=100)
    return x

data.index = data.index.map(fix_year)
data.head()
```

```python
# Step 5: Set the right dates as datetime64[ns]
data.index = pd.to_datetime(data.index)
print("\n Dates set as datetime64[ns]:")
data
```

```python
# Step 6: Compute the number of missing values for each location
missing_values_per_location = data.isnull().sum()
print("\n Missing values per location:")
missing_values_per_location
```

```python
# Step 7: Compute the number of non-missing values
non_missing_values_total = data.notnull().sum().sum()
print("\n Non-missing values in total:", non_missing_values_total)
```

```python
# Step 8: Calculate the mean windspeeds over all locations and times
mean_windspeed = data.mean().mean()
print("\n Mean windspeeds of the dataset:", mean_windspeed)
```

```python
# Step 9: Calculate min, max, mean, and standard deviation of windspeeds at each location
loc_stats = pd.DataFrame({
    'Min': data.min(),
    'Max': data.max(),
    'Mean': data.mean(),
    'Std': data.std()
})
print("\n Statistics for each location:")
loc_stats
```

```python
# Step 10: Calculate min, max, mean, and standard deviation of windspeeds across locations for each day
day_stats = pd.DataFrame({
    'Min': data.min(axis=1),
    'Max': data.max(axis=1),
    'Mean': data.mean(axis=1),
    'Std': data.std(axis=1)
})
print("Statistics for each day:")
day_stats

```

```python
# Step 11: Calculate average windspeed in January for each location
january_avg = data[data.index.month == 1].groupby(data[data.index.month == 1].index.to_period('Y')).mean()
print("\n Average windspeed in January for each location:")
january_avg
```

```python
# Steps 12-14: Downsample the record to yearly, monthly, and weekly frequencies
yearly = data.resample('Y').mean()
monthly = data.resample('M').mean()
weekly = data.resample('W').mean()
print("\n Downsampled to yearly, monthly and weekly frequency:")
yearly
monthly
weekly
```

```python
# Step 15: Calculate min, max, mean, and standard deviation of windspeeds across locations for each week for the first 52 weeks
weekly_stats = data.resample('W').agg(['min', 'max', 'mean', 'std']).iloc[:52]
print("\n Weekly statistics for the first 52 weeks:")
weekly_stats
```

### Question 6

Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.</br>
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

# Extracting data from the dataset
years = data['Year']
marriages_per_capita = data['Marriages_per_1000']
divorces_per_capita = data['Divorces_per_1000']

# Plotting the data
plt.figure(figsize=(10, 6))

plt.plot(years, marriages_per_capita, label='Marriages per Capita')
plt.plot(years, divorces_per_capita, label='Divorces per Capita')

plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867 - 2014)')
plt.legend()

plt.grid(True)
plt.show()
```

### Question 7

Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.</br>
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

# Filter data for the specific years (1900, 1950, and 2000)
years = [1900, 1950, 2000]
filtered_data = data[data['Year'].isin(years)]

# Extracting marriages and divorces per capita for the specific years
marriages = filtered_data[filtered_data['Year'].isin(years)]['Marriages_per_1000']
divorces = filtered_data[filtered_data['Year'].isin(years)]['Divorces_per_1000']

# Plotting the data in a vertical bar chart
plt.figure(figsize=(8, 6))

bar_width = 0.35
index = range(len(years))

plt.bar(index, marriages, bar_width, label='Marriages per Capita')
plt.bar([i + bar_width for i in index], divorces, bar_width, label='Divorces per Capita')

plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('Comparison of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')
plt.xticks([i + bar_width / 2 for i in index], years)
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```

### Question 8

Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort the actors by their kill count and label each bar with the corresponding actor's name. Don't forget to label your axes!</br>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('actor_kill_counts.csv')

# Sort the actors by their kill count in descending order
sorted_data = data.sort_values(by='Count', ascending=True)

# Create a horizontal bar chart
plt.figure(figsize=(10, 8))

plt.barh(sorted_data['Actor'], sorted_data['Count'], color='skyblue')
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood by Kill Count')

# Label each bar with the corresponding actor's name
for index, value in enumerate(sorted_data['Count']):
    plt.text(value, index, str(value))

plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

### Question 9

Create a pie chart showing the fraction of all Roman Emperors that were assassinated.</br>
Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.</br>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('roman-emperor-reigns.csv')

# Count the occurrences of assassination
assassinated_emperors = data[data['Cause_of_Death'] == 'Assassinated']
total_emperors = len(data)
total_assassinated = len(assassinated_emperors)

# Calculate the fraction of assassinated emperors
fraction_assassinated = total_assassinated / total_emperors

# Calculate the fraction of non-assassinated emperors
fraction_non_assassinated = 1 - fraction_assassinated

# Create a pie chart
labels = ['Assassinated', 'Non-assassinated']
sizes = [fraction_assassinated, fraction_non_assassinated]
explode = (0.1, 0)  # To explode the 'Assassinated' slice for emphasis

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Fraction of Roman Emperors Assassinated')

# Set aspect ratio to be equal to ensure a circular pie chart
plt.axis('equal')

plt.show()
```
